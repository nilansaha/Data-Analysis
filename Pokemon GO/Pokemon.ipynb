```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Read the dataset into a pandas dataframe and get rid of the pokemon index number
df = pd.read_csv('pokemonGO.csv')
df = df[['Name', 'Type 1', 'Type 2', 'Max CP', 'Max HP']]
```

```python
#Chart for the number of pokemons of each type
types = df['Type 1'].append(df['Type 2'])
types = types.dropna(how='any')
plt.figure()
sns.countplot(types, palette = "Set2")
plt.xticks(rotation = 45)
plt.title("Types of Pokemons")
```

```python
#Chart for the showing mean CP's for each type of pokemon
cp1 = df[['Type 1','Max CP']]
cp1 = cp1.rename(columns={'Type 1': 'Type'})
cp2 = df[['Type 2', 'Max CP']]
cp2 = cp2.rename(columns={'Type 2': 'Type'})
cp = cp1.append(cp2)
cp = cp.dropna(how="any")

plt.figure()
cp = cp.groupby('Type')['Max CP'].mean().reset_index()
cp = cp.sort_values('Max CP', ascending = False)
sns.barplot(x = 'Type',y = 'Max CP',data = cp, palette="Set2")
plt.xticks(rotation = 45)
plt.title("CP's according to Type")
```

```python
#Chart for the showing mean HP's for each type of pokemon
hp1 = df[['Type 1','Max HP']]
hp1 = hp1.rename(columns={'Type 1': 'Type'})
hp2 = df[['Type 2', 'Max HP']]
hp2 = hp2.rename(columns={'Type 2': 'Type'})
hp = hp1.append(hp2)
hp = hp.dropna(how="any")

plt.figure()
hp = hp.groupby('Type')['Max HP'].mean().reset_index()
hp = hp.sort_values('Max HP', ascending = False)
sns.barplot(x = 'Type', y = 'Max HP', data = hp, palette="Set2")
plt.xticks(rotation = 45)
plt.title("HP's according to Type")
```

```python
#Distribution of CP's
CP = df['Max CP']
plt.figure()
sns.distplot(CP, kde = False)
plt.title('Distribution of CP')
```

```python
#Distributiong of HP's
HP = df['Max HP']
plt.figure()
sns.distplot(HP, kde = False)
plt.title('Distribution of HP')
```
